```python
def clean_nse_data(df):
    # Clean column names
    nigeria_stock.columns = nigeria_stock.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('.', '')
    
    # Convert date to datetime and standardize format
    nigeria_stock['date'] = nigeria_stock.to_datetime(df['date'], format='%d/%m/%Y')
    
    # Clean numeric columns (remove commas, % signs)
    numeric_cols = ['price', 'open', 'high', 'low', 'change_%']
    
    for col in numeric_cols:
        # First ensure the column is string type
        nigeria_stock[col] = nigeria_stock[col].astype(str)
        
        if col == 'change_%':
            nigeria_stock[col] = nigeria_stock[col].str.replace('%', '').astype(float) / 100
        else:
            nigeria_stock[col] = nigeria_stock[col].str.replace(',', '').astype(float)
    
    # Clean volume column
    def clean_volume(vol):
        if pd.isna(vol):
            return np.nan
        vol = str(vol)  # Ensure it's string
        if 'M' in vol:
            return float(vol.replace('M', '')) * 1_000_000
        elif 'B' in vol:
            return float(vol.replace('B', '')) * 1_000_000_000
        elif 'K' in vol:
            return float(vol.replace('K', '')) * 1_000
        else:
            return float(vol)
    
    nigeria_stock['vol'] = nigeria_stock['vol'].apply(clean_volume)
    
    # Rename columns for consistency
    nigeria_stock = nigeria_stock.new(columns={
        'change_%': 'change_pct',
        'vol': 'volume'
    })
    
    # Reorder columns
    nigeria_stock = nigeria_stock[['date', 'price', 'open', 'high', 'low', 'volume', 'change_pct']]
    
    # Sort by date
    nigeria_stock = nigeria_stock.sort_values('date')
    
    return nigeria_stock
```

```python
def convert_volume(vol):
    """
    Convert volume strings with M/B suffixes to numeric values
    Examples:
    "423.20M" → 423200000
    "1.03B" → 1030000000
    """
    if pd.isna(vol):
        return vol
    vol = str(vol).upper().strip()  # Ensure string and uppercase
    
    if 'M' in vol:
        return float(vol.replace('M', '')) * 1_000_000
    elif 'B' in vol:
        return float(vol.replace('B', '')) * 1_000_000_000
    elif 'K' in vol:
        return float(vol.replace('K', '')) * 1_000
    else:  # If no suffix, assume raw number
        return float(vol)
```

```python
# Apply the conversion
nigeria_stock['volume'] = nigeria_stock['volume'].apply(convert_volume)
```
